```python
import os
import json

# File to save tasks
FILE_NAME = "tasks.json"

def load_tasks():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, 'r') as file:
            return json.load(file)
    return []

def save_tasks(tasks):
    with open(FILE_NAME, 'w') as file:
        json.dump(tasks, file)

def add_task(tasks):
    task_name = input("Enter task name: ")
    deadline = input(int("Enter deadline (e.g., 2024-08-15): "))
    priority = input("Enter priority (High/Medium/Low): ")
    tasks.append({
        'task': task_name,
        'deadline': deadline,
        'priority': priority,
        'completed': False
    })
    print("Task added successfully!")

def remove_task(tasks):
    task_id = int(input("Enter task ID to remove: ")) - 1
    if 0 <= task_id < len(tasks):
        removed_task = tasks.pop(task_id)
        print(f"Task '{removed_task['task']}' removed successfully!")
    else:
        print("Invalid task ID.")

def mark_task_complete(tasks):
    task_id = int(input("Enter task ID to mark as complete: ")) - 1
    if 0 <= task_id < len(tasks):
        tasks[task_id]['completed'] = True
        print(f"Task '{tasks[task_id]['task']}' marked as complete!")
    else:
        print("Invalid task ID.")

def display_tasks(tasks):
    if not tasks:
        print("No tasks to display.")
    else:
        for i, task in enumerate(tasks, 1):
            status = "Completed" if task['completed'] else "Incomplete"
            print(f"{i}. {task['task']} - {task['deadline']} - {task['priority']} - {status}")

def main():
    tasks = load_tasks()
    while True:
        print("\n1. Add Task\n2. Remove Task\n3. Mark Task as Complete\n4. Display Tasks\n5. Save & Exit")
        choice = input("Choose an option: ")
        if choice == '1':
            add_task(tasks)
        elif choice == '2':
            display_tasks(tasks)
            remove_task(tasks)
        elif choice == '3':
            display_tasks(tasks)
            mark_task_complete(tasks)
        elif choice == '4':
            display_tasks(tasks)
        elif choice == '5':
            save_tasks(tasks)
            print("Tasks saved. Exiting the application.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
