```python
# Library Management System
users = {}  # Store user credentials
books = {}  # Store book details

def register():
    username = input('Register - Enter a new username: ')
    if username in users:
        print("Username already exists. Try again.")
        return
    password = input('Enter a new password: ')
    users[username] = password
    print("Registration successful!")

def login():
    if not users:
        print("No registered users. Please register first.")
        return False
    username = input('Login - Enter your username: ')
    password = input('Enter your password: ')
    if username in users and users[username] == password:
        print("Login successful!")
        return True
    else:
        print("Invalid credentials. Try again.")
        return False

#crud operation
def add_books():
    book_id = input('Enter book ID: ')
    name = input('Enter book name: ')
    author = input('Enter the author name: ')
    books[book_id] = {'book_name': name, 'book_author': author}
    print("Book added successfully!")

def display_books():
    if not books:
        print("No records to display")
    else:
        print("\nBook Records")
        for book_id, details in books.items():
            print(f'\nBook ID: {book_id}')
            print(f'Book Name: {details["book_name"]}')
            print(f'Book Author: {details["book_author"]}')
            print("_" * 10)

def search_books():
    book_id = input("Enter the book ID you want to search: ")
    if book_id in books:
        print("\nBook Record")
        print(f'Book ID: {book_id}')
        print(f'Book Name: {books[book_id]["book_name"]}')
        print(f'Book Author: {books[book_id]["book_author"]}')
    else:
        print("No records found")

def remove_books():
    book_id = input("Enter the book ID you want to delete: ")
    if book_id in books:
        books.pop(book_id)
        print("Book details deleted successfully")
    else:
        print("No records found")

def main():
    while True:
        choices = input("\nChoose an option: \n1. Register \n2. Login \n3. Exit\n")
        if choices == '1':
            register()
        elif choices == '2':
            if login():
                while True:
                    choice = input("\nEnter your choice: \n1. Add Book \n2. Display Books \n3. Search Book \n4. Remove Book \n5. Logout \n")
                    if choice == '1':
                        add_books()
                    elif choice == '2':
                        display_books()
                    elif choice == '3':
                        search_books()
                    elif choice == '4':
                        remove_books()
                    elif choice == '5':
                        print("Logging out...")
                        break
                    else:
                        print("Enter a valid choice.")
        elif choices == '3':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
