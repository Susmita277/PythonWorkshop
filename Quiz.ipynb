```python
import time

# Sample quiz questions
questions = [
    {
        "question": "What is 2+8?",
        "options": ["1. 10", "2. 13", "3. 9", "4. 15"],
        "answer": 1
    },
    {
        "question": "Who is the father of C?",
        "options": ["1.Guido van Rossum", "2.Bjarne Stroustrup", "3.James Gosling", "4.Dennis Ritchie"],
        "answer": 4
    },
  
]

def ask_question(question_data):
    print("\n" + question_data["question"])
    for option in question_data["options"]:
        print(option)
    
    start_time = time.time()
    try:
        answer = int(input("Enter the correct option number: "))
    except ValueError:
        print("Invalid input! Please enter a number.")
        return False, 0
    
    end_time = time.time()
    elapsed_time = end_time - start_time

    if answer == question_data["answer"]:
        print(f"Correct! You took {elapsed_time:.2f} seconds.")
        return True, elapsed_time
    else:
        print(f"Wrong answer. The correct answer was {question_data['answer']}. You took {elapsed_time:.2f} seconds.")
        return False, elapsed_time

def main():
    score = 0
    total_time = 0

    for question_data in questions:
        correct, elapsed_time = ask_question(question_data)
        total_time += elapsed_time
        if correct:
            score += 1
    
    print(f"\nQuiz completed! Your final score is {score} out of {len(questions)}.")
    print(f"Total time taken: {total_time:.2f} seconds.")

if __name__ == "__main__":
    main()
```
